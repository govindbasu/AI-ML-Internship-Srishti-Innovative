```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]

plt.plot(x, y)

plt.title("Line Graph")
plt.xlabel("X Values")
plt.ylabel("Y Values")

plt.show()
```

```python
import matplotlib.pyplot as plt

students = ["Arun", "Rahul", "Neha", "Diya"]
marks = [85, 90, 78, 88]

plt.bar(students, marks)

plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```

```python
import matplotlib.pyplot as plt

data = [40, 35, 25]
labels = ["Python", "Java", "C++"]

plt.pie(data, labels=labels)

plt.title("Programming Language Usage")

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [5,7,6,8,10]

plt.scatter(x, y)

plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

```python
import matplotlib.pyplot as plt

data = [10,10,20,20,20,30,30,40,50,50,50,50]

plt.hist(data, bins=[0,10,20,30,40,50,60], edgecolor="black")

plt.title("Frequency Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]

y1 = [10,20,30,40]
y2 = [40,30,20,10]

plt.plot(x, y1, label="Sales")
plt.plot(x, y2, label="Loss")

plt.legend()

plt.title("Multiple Lines")

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,15,25]

plt.plot(x, y, linestyle="--", marker="o")

plt.title("Styled Graph")

plt.show()
```

```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5]
temp = [30,32,31,35,33]

plt.plot(days, temp, marker="o")

plt.title("Temperature Analysis")
plt.xlabel("Days")
plt.ylabel("Temperature")

plt.show()
```
